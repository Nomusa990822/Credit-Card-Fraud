# Credit Card Fraud Detection — Model Optimization and Threshold Tuning

## Project Overview

In Project 2, baseline machine learning models were developed to classify transactions as fraudulent or legitimate. The main findings were:

- Logistic Regression provided a useful baseline but produced weak fraud-class precision
- Random Forest improved the balance between precision and recall
- XGBoost achieved the strongest overall ranking performance

However, baseline performance is not enough for a strong fraud detection system. In real financial applications, the classification threshold and hyperparameter configuration must be optimized carefully to reduce false positives while still detecting as much fraud as possible.

## Objective

The goal of this project is to improve fraud detection performance through:

1. hyperparameter tuning
2. class imbalance handling
3. threshold tuning
4. precision-recall trade-off analysis
5. final optimized model selection

## Why This Matters

Fraud detection is a high-stakes classification problem. A model that detects more fraud may also create more false alarms, while a model that is too conservative may miss costly fraudulent transactions. This project focuses on finding a more practical balance.
